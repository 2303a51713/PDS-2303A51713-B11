```python
#Q1
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("population.csv")
top5 = df.sort_values(by="Population", ascending=False).head(5)
plt.figure(figsize=(8,6))
bars = plt.bar(top5["Country"], top5["Population"], color="skyblue", edgecolor="black")
plt.ylabel("Population (in Billions)", fontsize=12)
plt.xlabel("Country", fontsize=12)
plt.title("Top 5 Countries by Population", fontsize=14)
def billions(x, pos):
    return f'{x/1e9:.1f}B'

from matplotlib.ticker import FuncFormatter
plt.gca().yaxis.set_major_formatter(FuncFormatter(billions))

for bar in bars:
    plt.text(bar.get_x() + bar.get_width()/2,
             bar.get_height(),
             f'{bar.get_height()/1e9:.2f}B',
             ha='center', va='bottom', fontsize=10)

plt.tight_layout()
plt.show()

largest_country = top5.iloc[0]["Country"]
print(f"The country with the largest population is: {largest_country}")
```

```python
#Q2
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("movies_genre.csv")
df = df.sort_values(by="Movie_Count", ascending=False)
plt.figure(figsize=(10,6))
bars = plt.bar(df["Genre"], df["Movie_Count"], color="lightgreen", edgecolor="black")

plt.xlabel("Movie Genres", fontsize=12)
plt.ylabel("Number of Movies", fontsize=12)
plt.title("Number of Movies Produced by Genre", fontsize=14)
plt.xticks(rotation=45, ha="right")
for bar in bars:
    plt.text(bar.get_x() + bar.get_width()/2,
             bar.get_height(),
             str(int(bar.get_height())),
             ha='center', va='bottom', fontsize=9)

plt.tight_layout()
plt.show()
most_popular_genre = df.iloc[0]["Genre"]
print(f"The most popular movie genre is: {most_popular_genre}")
```
